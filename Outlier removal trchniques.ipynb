```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('HeightWeight.csv')
```

```python
sns.histplot(df['Weight(Pounds)'], kde=True)
```

```python
mean_value = df['Weight(Pounds)'].mean()
mean_value
```

```python
std_value = df['Weight(Pounds)'].std()
std_value
```

```python
std_3 = mean_value + (3 * std_value)
std_3
```

```python
neg_std_3 = mean_value - (3 * std_value)
```

```python
df[df['Weight(Pounds)'] > std_3]
```

```python
df[df['Weight(Pounds)'] < neg_std_3]
```

```python
merge = df[(df['Weight(Pounds)'] > std_3) | (df['Weight(Pounds)'] < neg_std_3)]
merge
```

```python
merge.shape
```

```python
#Removing outlier
ro = df[(df['Weight(Pounds)'] < std_3) & (df['Weight(Pounds)'] > neg_std_3)]
ro.shape
```

```python
sns.histplot(ro['Weight(Pounds)'], kde=True)
```
